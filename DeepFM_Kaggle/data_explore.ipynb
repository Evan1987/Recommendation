```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from data_utils import load_data


sns.set_style("whitegrid")
sns.set_palette("Set3")
```

```python
data_alias = load_data()
train_data, test_data = data_alias["train"], data_alias["test"]
```

```python
print(f"Train: {train_data.shape}")
print(f"Test: {test_data.shape}")
```

```python
train_data["target"].value_counts()
```

```python
# ... check for -1 (NAs in our data)
train_nulls = (train_data == -1).sum() / len(train_data)
train_nulls[train_nulls > 0].sort_values(ascending=False)
```

```python
test_nulls = (test_data == -1).sum() / len(test_data)
test_nulls[test_nulls > 0].sort_values(ascending=False)
```

```python
all_features = train_data.columns.difference(["id", "target"])  # 57
numeric_features, categorical_features, binary_features = [], [], []
for col in all_features:
    target = col[-3:]
    if target == "bin":
        binary_features.append(col)
    elif target == "cat":
        categorical_features.append(col)
    else:
        numeric_features.append(col)

print(f"Num of numeric features: {len(numeric_features)}")  # 26
print(f"Num of binary features: {len(binary_features)}")  # 17
print(f"Num of categorical features: {len(categorical_features)}")  # 14
```

```python
corr = train_data[numeric_features].corr(method="pearson").round(2)
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(h_neg=250, h_pos=10, as_cmap=True)
fig, axe = plt.subplots(figsize=(20, 20))
sns.heatmap(corr, annot=True, mask=mask, ax=axe)
fig.show()
```

## Numeric Features Exploration

```python
true_cond = train_data["target"] == 1
for column in numeric_features:
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))
    
    fig.suptitle(f"Plots for {column}", fontsize=18)
    
    # Distribution plot
    ax = axes[0, 0]
    sns.distplot(train_data[column], ax=ax)
    ax.set_xlabel(column, fontsize=14)
    ax.set_ylabel("Density", fontsize=14)
      
    # Distribution on labels
    ax = axes[0, 1]
    sns.distplot(train_data.loc[true_cond, column], color="red", label="Claim not filed", ax=ax)
    sns.distplot(train_data.loc[~true_cond, column], color="blue", label="Claim filed", ax=ax)
    ax.legend(loc="best")
    ax.set_xlabel(column, fontsize=14)
    ax.set_ylabel("Density", fontsize=14)
    
    # Average column value per label
    ax = axes[1, 0]
    sns.barplot(x="target", y=column, data=train_data, ax=ax)
    ax.set_xlabel("Is Filed Claim?", fontsize=14)
    ax.set_ylabel(column, fontsize=14)
    
    # Boxplot
    ax = axes[1, 1]
    sns.boxplot(x="target", y=column, data=train_data, ax=ax)
    ax.set_xlabel("Is Filed Claim?", fontsize=14)
    ax.set_ylabel(column, fontsize=14)
    
    fig.show()
```

## Categorical Features Exploration

```python
def plot_for_categorical(column):
    fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(18, 12))
    fig.suptitle(f"Plots for {column}", fontsize=18)
    
    # Number of occurrences per category-target pair
    ax = axes[0]
    sns.countplot(x=column, hue="target", data=train_data, ax=ax)
    ax.set_xlabel(column, fontsize=14)
    ax.set_ylabel("Number of Occurrences", fontsize=14)

    height = [p.get_height() for p in ax.patches]  # because of hue, it's #hue size of ncols
    ncol = int(len(height) / 2)
    total = [height[i] + height[i + ncol] for i in range(ncol)] * 2
    # add label ticks on each bar
    for i, p in enumerate(ax.patches):
        ax.text(x=p.get_x() + p.get_width() / 2, 
                y=height[i] * 1.01 + 1000, 
                s=f"{height[i] / total[i]:.0%}", 
                ha="center", 
                size=14)
    
    ax = axes[1]
    sns.pointplot(x=column, y="target", data=train_data, ax=ax)
    ax.set_xlabel(column, fontsize=14)
    ax.set_ylabel("Filed Claims Percentage", fontsize=14)
    
    fig.show()
```

```python
for column in categorical_features:
    plot_for_categorical(column)
```

## Binary Features Exploration

```python
for column in binary_features:
    plot_for_categorical(column)
```
